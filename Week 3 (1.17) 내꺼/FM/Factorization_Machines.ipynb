Factorization Machine은 linear regression과 동일한데 linear regression의 한계점인 변수간의 교호작용을 고려하지 못한다는 단점을 커버한 모델이다. 

다운로드 (1).png

위에서 보여지듯이 단순히 wx(기존의 linear regression)이 아니라 xi와 xj의 곱, 즉 변수간의 교호작용을 고려한다는 것을 볼 수 있다. 이떄 j = i + 1인데 이는 i와 j가 같지 않게 설정한다는 뜻이고 (i,j)와 (j,i)처럼 revers로 반복적인 계산은 하지 않겠다는 뜻이다. W = <vi,vj>가 symmetric matrix로 나오는데 symmetric matrix는 QΛQT로 변환할 수 있다. 그리고 이는 K.transpoe()K로 나타낼 수 있다. 즉 n*k의 matrix를 가지게 되는데 이를 통해 linear complexity를 가진다는 것을 알 수 있다.
